## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """select * from master 
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`



```python
%pip install seaborn

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
Query='''select * from master'''
df=pd.read_sql_query(Query,conn)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(df['ConvertedComp'], bins=120, kde=False)

# Axis x
formatter = mticker.FuncFormatter(lambda x, pos: f"{x:,.0f}")
plt.gca().xaxis.set_major_formatter(formatter)
plt.xticks(rotation=75)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
sns.boxplot(y="Age", data=df)
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
sns.scatterplot(x="Age",y="WorkWeekHrs",data=df)
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
sns.scatterplot(x="CodeRevHrs", y="WorkWeekHrs", size="Age", sizes=(20, 200), data=df, legend=False, color='blue', edgecolor='black')


plt.xlabel('CodeRevHrs')
plt.ylabel('WorkWeekHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
Query='''select * from DatabaseDesireNextYear'''##First we bring the sql here
df=pd.read_sql_query(Query,conn)
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Count the occurrences of each database
database_counts = df['DatabaseDesireNextYear'].value_counts()

# Get the top 5 databases
top_databases = database_counts.nlargest(5)

# Create the pie chart
plt.figure(figsize=(10, 6))
plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=140)

# Add title
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular

# Show the pie chart
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
Query='''select * from master'''
df=pd.read_sql_query(Query,conn)
```

```python
df2=df[(df['Age']>=30)&(df['Age']<=35)]
df3=df2.groupby('Age')[['WorkWeekHrs','CodeRevHrs']].median()
df3.plot(kind='bar', stacked=True)
plt.xticks(rotation=0)
plt.legend(loc='lower right')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calcula la mediana de ConvertedComp por edad
median_comp = filtered_df.groupby('Age')['ConvertedComp'].median().reset_index()

# Crea el gráfico
plt.figure(figsize=(10, 6))
sns.lineplot(x='Age', y='ConvertedComp', data=median_comp, palette='viridis')

# Añade título y etiquetas
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')

# Muestra el gráfico
plt.show()

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# Count the occurrences of each category in MainBranch
main_branch_counts = df['MainBranch'].value_counts()

# Create the horizontal bar chart
plt.figure(figsize=(7, 4))
sns.barplot(x=main_branch_counts.values, y=main_branch_counts.index, palette='viridis')

# Add title and labels
plt.title('Count of Main Branches')
plt.xlabel('Count')
plt.ylabel('Main Branch')

# Show the chart
plt.show()
```
